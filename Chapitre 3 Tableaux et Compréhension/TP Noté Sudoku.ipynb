### 1) Faire le tableau de 3 lignes et 3 colonnes suivant en python (il y a 1,2,3 sur la première ligne, puis 4,5,6 sur la seconde et finalement 7,8 et 9 sur la dernière ligne). Le tableau s'appelera M.

### 9) Faire une fonction qui prend en entrée un tableau qui affiche valeur par valeur toutes les valeurs du tableau. (différent de AfficherTableau qui affiche les valeurs 3 par 3)

```python
M = [[1,2,3],[4,5,6],[7,8,9]]

def afficher(M):
    for ligne in range(len(M)):
        for colonne in M[ligne]:
            print(colonne)

afficher(M)
```
